```python
import pandas as pd
```

```python
df = pd.read_excel("teste.xlsx")
df.head(10)
```

```python
pivot_table = pd.pivot_table(
    data=df,
    values='VALOR',     # coluna com os valores numéricos a serem agregados
    index='CLIENTE',         # coluna(s) para usar como índice na nova tabela
    columns='Mes_Ano',   # Usa a nova coluna 'Mes_Ano'
    aggfunc='sum',
    margins=True,        # Adiciona o total geral
    margins_name='Total Geral'  # Nomeia o total geral
)
pivot_table.fillna(0, inplace=True)

pivot_table.columns = pivot_table.columns.strftime('%b/%y').str.upper()
# Função para formatar valores negativos em vermelho
def color_negative_red(val):
    color = 'red' if val < 0 else 'black'
    return f'color: {color}'

# Aplicar o estilo na tabela pivô usando `map`
tabela_pivo_style = pivot_table.style.map(color_negative_red)

tabela_pivo_style
```
